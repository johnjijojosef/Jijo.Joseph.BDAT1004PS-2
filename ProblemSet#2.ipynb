#  Question 3:
Write a class named Marsupial that can be used as shown below:
    
>>> m = Marsupial()
>>> m.put_in_pouch('doll')
>>> m.put_in_pouch('firetruck')
>>> m.put_in_pouch('kitten')
>>> m.pouch_contents()
['doll', 'firetruck', 'kitten']

Now write a class named Kangaroo as a subclass of Marsupial that inherits all the 
attributes of Marsupial and also:
a. extends the Marsupial __init__ constructor to take, as input, the 
coordinates x and y of the Kangaroo object,
b. supports method jump that takes number values dx and dy as input and 
moves the kangaroo by dx units along the x-axis and by dy units along the yaxis, and
c. overloads the __str__ operator so it behaves as shown below.
>>> k = Kangaroo(0,0)
>>> print(k)
I am a Kangaroo located at coordinates (0,0)
>>> k.put_in_pouch('doll')
>>> k.put_in_pouch('firetruck')
>>> k.put_in_pouch('kitten')
>>> k.pouch_contents()
['doll', 'firetruck', 'kitten']
>>> k.jump(1,0)
>>> k.jump(1,0)
>>> k.jump(1,0)
>>> print(k)
I am a Kangaroo located at coordinates (3,0)

```python
class Marsupial:

    def __init__(self):
        # Empty list is initialized
        self.pouch = []

    # Function put_in_pouch to add the items to the list
    def put_in_pouch(self, item):
        self.pouch.append(item)

    # Function to display the contents in the pouch
    def pouch_contents(self):
        return self.pouch

#Subclass of Marsupial

class Kangaroo(Marsupial):
    # a and b are variables to store the coordinate values
    def __init__(self, a, b):
        # To call the parent class
        Marsupial.__init__(self)
        self.x = a
        self.y = b

    # Function to store and increase the values where c and d are the values being displaced upon calling the function
    def jump(self, c, d):
        self.x = self.x + c
        self.y = self.y + d

    # String function to display the output jumps from the jump function
    def __str__(self):
        return "I am a Kangaroo located at coordinates (" + str(self.x) + ", " + str(self.y) +")"


# An instance for class kangaroo is created initialized with (0,0) as coordinates
k = Kangaroo(0, 0)

# To print the present coordinates
print(k)

# By using the instance variable of kangaroo class we are calling Marsupial class
k.put_in_pouch("doll")
k.put_in_pouch("firetruck")
k.put_in_pouch("kitten")

# Displaying the list of contents in the Marsupial
print(k.pouch_contents())

#Calling jump function using the instance
k.jump(1, 0)
k.jump(1, 0)
k.jump(1, 0)

# To print the final displacement of from coordinate (0,0)
print(k)
```

# # Question 6

Implement a class named HeadingParser that can be used to parse an HTML document, and retrieve and print all the headings in the document. You should implement your class as a subclass of HTMLParser, defined in Standard Library module html.parser. When fed a string containing HTML code, your class should print the headings, one per line and in the order in which they appear in the document. Each heading should be indented as follows: an h1 heading should have indentation 0, and h2 heading should have indentation 1, etc. Test your implementation using w3c.html.

>>> infile = open('w3c.html')
>>> content = infile.read()
>>> infile.close()
>>> hp = HeadingParser()
>>> hp.feed(content)
W3C Mission
Principles

Reference :https://docs.python.org/3/library/html.parser.html

```python
from html.parser import HTMLParser

class HeadingParser(HTMLParser):
    head1 = False   # Flag to indicate if inside an <h1> tag
    head2 = False   # Flag to indicate if inside an <h2> tag
    head3 = False   # Flag to indicate if inside an <h3> tag
    head4 = False   # Flag to indicate if inside an <h4> tag
    
#Function to check for headers using start tags in html for h1 , h2 , h3 and h4
    def handle_starttag(self, tag, attrs):
        if tag == "h1":
            self.head1 = True
        elif tag == "h2":
            self.head2 = True
        elif tag == "h3":
            self.head3 = True
        elif tag == "h4":
            self.head4 = True 
#Function to print the data that was retrieved in the handle_starttag
    def handle_data(self, data):
        if self.head1:
            print("Heading 1:", data)
        elif self.head2:
            print("Heading 2:", data)
        elif self.head3:
            print("Heading 3:", data)   
        elif self.head4:
            print("Heading 4:", data)   
            
#Function to check for headers using end tags in html for h1 , h2 , h3 and h4

    def handle_endtag(self, tag):
        if tag == "h1":
            self.head1 = False
        elif tag == "h2":
            self.head2 = False
        elif tag == "h3":
            self.head3 = False  
        elif tag == "h4":
            self.head3 = False  
#Creating an object for the class           
hParser = HeadingParser()
#Open the file using read mode
file = open("w3c.html", "r")
#Read the file and store in html variable
html = file.read()
# Close the file
file.close()
# Pass the file to the class
hParser.feed(html)
```

Implement recursive function webdir() that takes as input: a URL (as a string) and non-negative integers depth and indent. Your function should visit every web page reachable from the starting URL web page in depth clicks or less, and print each web page's URL. As shown below, indentation, specified by indent, should be used to indicate the depth of a URL.

>>> 
webdir('http://reed.cs.depaul.edu/lperkovic/csc242/test1.html'
, 2, 0)
http://reed.cs.depaul.edu/lperkovic/csc242/test1.html
 http://reed.cs.depaul.edu/lperkovic/csc242/test2.html
 http://reed.cs.depaul.edu/lperkovic/csc242/test4.html
 http://reed.cs.depaul.edu/lperkovic/csc242/test3.html
 http://reed.cs.depaul.edu/lperkovic/csc242/test4.html
 
 Reference :https://stackoverflow.com/questions/67293733/how-to-extract-all-links-from-a-website-using-python
 


```python
from urllib.request import urlopen
from urllib.parse import urljoin
from bs4 import BeautifulSoup

#Function to retrieve the links that are in a specific URL
def webdir(url, depth):
#To print the URL which is currently being checked
    print(url)  

#When the depth reaches 0 the execution will stop
    if depth == 0:
           return 
#To open the URL and get the response
    response = urlopen(url)
#To decode the reponse and get the html content
    dec = response.read().decode()
#To create an object to retrieve the contentss from HTML
    soup = BeautifulSoup(dec, 'html.parser')  # Create a BeautifulSoup object for parsing the HTML
#The tag a[href] is will be selected and stored in variable anchor
    x = soup.select('a[href]')  

    for m in x:
        ref = m['href'] # To get the href values # Get the value of the href attribute
        ab = urljoin(url, ref)  # The anchor urls are being joined with the base URL

        # Function calling recurssively subtracting the depth by -1
        webdir(ab, depth - 1)

# Invoking the function to scrape the url
webdir('http://reed.cs.depaul.edu/lperkovic/test1.html', 2)
```

Write list comprehension expressions that use list words and generate the following 
lists:

a) ['THE', 'QUICK', 'BROWN', 'FOX', 'JUMPS', 'OVER', 'THE', 'LAZY', 'DOG']

b) ['the', 'quick', 'brown', 'fox', 'jumps', 'over', 'the', 'lazy', 'dog']

```python
# Creating a list with the words
words = ['The', 'quick', 'brown', 'fox', 'jumps', 'over', 'the', 'lazy', 'dog']

# Listing all words upper case

upper_case=[]
for m in words:
    a= m.upper()
    upper_case.append(a)
print("Answer 1:",upper_case)

# Listing all words lower case

lower_case=[]
for m in words:
    a= m.lower()
    lower_case.append(a)
print("Answer 2:",lower_case)

# Listing length of each word
length=[]
for m in words:
    a=len(m)
    length.append(a)

print("Answer 3:",length)

# Listing the list containing a list for every word of list words, where each list contains the word in uppercase and lowercase and the length of the word
upper_lower_length = []
for m in words:
    a= m.upper()
    b= m.lower()
    c= len(m)
    d = a ,b ,c
    upper_lower_length.append(d)

print("Answer 4:",upper_lower_length)

# Listing the list of words in list words containing 4 or more characters
length_list=[]
for m in words:
    if len(m)>=4:
        length_list.append(m)
        
print("Answer 5:",length_list)
```
